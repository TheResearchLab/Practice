```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of the discrete random variable
x_values = np.arange(1, 7)

# Define the PMF for a fair six-sided die
pmf_values = np.ones_like(x_values) / len(x_values)

# Plot the PMF
plt.bar(x_values, pmf_values, tick_label=x_values)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass Function of a Fair Six-Sided Die')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data for a normal distribution (mean=0, standard deviation=1)
x_values = np.linspace(-5, 5, 1000)  # Range of x values
pdf_values = norm.pdf(x_values, loc=0, scale=1)  # PDF values for the normal distribution

# Plot the PDF
plt.plot(x_values, pdf_values, label='PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of a Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data for a normal distribution 
x_values = np.linspace(-5, 5, 1000)
cdf_values = norm.cdf(x_values, loc=0, scale=1)

# Plot the CDF
plt.plot(x_values, cdf_values, label='CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) of a Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

# Poisson Distribution

$$ \mu = \lambda $$

$$ \sigma^2 = \lambda $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
lmbda = 5
size = 10000

# Generate random numbers from Poisson distribution
data = np.random.poisson(lmbda, size)

# Plot the histogram
plt.hist(data, bins=20, density=True, color='skyblue', edgecolor='black')

# Set labels and title
plt.xlabel('Value')
plt.ylabel('Probability Mass')
plt.title('Poisson Distribution')

# Show plot
plt.show()
```

# Exponential Distribution

$$
\mu=\frac{1}{\lambda}
$$

$$
\sigma^2 = \frac{1}{\lambda^2}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

lmbda, size = 0.5, 1000
data = np.random.exponential(1/lmbda, size)
plt.hist(data, bins=20, density=True, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution')
plt.show()
```

# Hypergeometric Distribution

$$
\mu = \frac{nK}{N}
$$

$$
\sigma^2 = n \frac{K}{N} \frac{N-K}{N} \frac{N-n}{N-1}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
N = 20  # Population size
K = 8   # Number of successes in the population
n = 5   # Number of draws (samples) without replacement
size = 1000  # Number of samples

# Generate random numbers from hypergeometric distribution
data = np.random.hypergeometric(K, N-K, n, size)

# Plot the histogram
(plt.hist(data, 
          bins=np.arange(min(data), 
          max(data)+1.5)-0.5, 
          density=True, 
          color='skyblue', 
          edgecolor='black')
)
# Set labels and title
plt.xlabel('Value')
plt.ylabel('Probability Mass')
plt.title('Hypergeometric Distribution')

# Show plot
plt.show()
```

# Normal Distribution

$$
f(x | \mu, \sigma^2) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


mean = 0  
std_dev = 1  
data = np.random.normal(mean, std_dev, 1000)


plt.hist(data, bins=30, density=True, alpha=0.6, color='g')


xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

# Standardization

$$
z = \frac{x - \mu}{\sigma}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a normally distributed dataset
np.random.seed(0)
mu = 50  # mean
sigma = 10  # standard deviation
data = np.random.normal(mu, sigma, 1000)

# Standardize the dataset
z_scores = (data - np.mean(data)) / np.std(data)

# Plot the original and standardized data
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', alpha=0.7)
plt.title('Original Data (Normally Distributed)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='salmon', alpha=0.7)
plt.title('Standardized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
