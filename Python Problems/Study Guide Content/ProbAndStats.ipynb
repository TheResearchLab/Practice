```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of the discrete random variable
x_values = np.arange(1, 7)

# Define the PMF for a fair six-sided die
pmf_values = np.ones_like(x_values) / len(x_values)

# Plot the PMF
plt.bar(x_values, pmf_values, tick_label=x_values)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass Function of a Fair Six-Sided Die')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data for a normal distribution (mean=0, standard deviation=1)
x_values = np.linspace(-5, 5, 1000)  # Range of x values
pdf_values = norm.pdf(x_values, loc=0, scale=1)  # PDF values for the normal distribution

# Plot the PDF
plt.plot(x_values, pdf_values, label='PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF) of a Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data for a normal distribution 
x_values = np.linspace(-5, 5, 1000)
cdf_values = norm.cdf(x_values, loc=0, scale=1)

# Plot the CDF
plt.plot(x_values, cdf_values, label='CDF')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) of a Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
